# Water Quality Detection Process
## 1.Data Collection

The dataset for water quality detection contains chemical measurements collected from different water sources. Each row represents a water sample, and each column represents a specific attribute of the water quality.

The target column Potability indicates whether the water is potable (1) or not potable (0).

## 2.Data Preprocessing

Before training the models, the data must be preprocessed. This includes handling missing values, encoding categorical variables, scaling numerical features, and splitting the data into training and testing sets.
## 3.Feature Selection
The following features are used for model training:
- **pH**: Indicates the acidity or alkalinity of the water.
- **Hardness**: Measures the concentration of calcium and magnesium ions.
- **Solids**: Total dissolved solids in water.
- **Chloramines**: Measures the concentration of chloramines.
- **Sulfate**: Indicates sulfate ion concentration.
- **Conductivity**: Electrical conductivity in the water sample.
- **Organic Carbon**: Measures the amount of organic carbon present.
- **Trihalomethanes**: Concentration of trihalomethanes, byproducts of chlorination.
- **Turbidity**: The cloudiness or haziness of the water.
- **Potability**: The target column, where 1 indicates potable water and 0 indicates non-potable.
## 4.Model Training

 The following machine learning models are trained to predict whether the water is potable or not:
 - Support Vector Machine (SVM): A popular classification algorithm.
 - XGBoost: A gradient-boosting algorithm known for handling structured data well.

## 5.Model Evaluation

The performance of each model is evaluated using accuracy scores. The results for each model are:
- SVM: 68% accuracy.
- XGBoost: 67% accuracy.
## 6.Best Model Selection

In this case, the SVM model performs slightly better than XGBoost with an accuracy of 68%. However, neither model achieves extremely high accuracy, indicating that further tuning or feature engineering might be required to improve performance.
## Columns in the Dataset
Here is a detailed description of the columns used in the dataset:

- **pH**: Acidity or alkalinity of the water.
- **Hardness**: The concentration of calcium and magnesium ions.
- **Solids**: The total dissolved solids in mg/L.
- **Chloramines**: Concentration of chloramines in ppm.
- **Sulfate**: Sulfate ion concentration in mg/L.
- **Conductivity**: Electrical conductivity in μS/cm.
- **Organic Carbon**: The amount of organic carbon in ppm.
- **Trihalomethanes**: Concentration of trihalomethanes in μg/L.
- **Turbidity**: Cloudiness or haziness of the water.
- **Potability**: Target variable (1 = potable, 0 = not potable).

By preprocessing this data, selecting the relevant features, and training models like SVM and XGBoost, it is possible to predict the potability of water. While SVM shows slightly better performance, further model optimization may lead to better results in the future.
